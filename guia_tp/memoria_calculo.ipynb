```python
import pandas as pd
import numpy as np
import plotly.express as px
from scipy.constants import Planck, elementary_charge, speed_of_light
```

```python
def calcular_energia_foton(longitud_nm):
    return (Planck * speed_of_light / (longitud_nm * 1e-9)) / elementary_charge
```

```python
df = pd.read_csv("data_2023.csv")
df["U_media"] = df[["G1", "G2", "G3"]].mean(axis=1).round(3)
df.sort_values("U_media", ascending=False, inplace=True)
df["E_foton[eV]"] = calcular_energia_foton(df["long [nm]"]).round(3)
df["Ua"] = df["E_foton[eV]"]  - df["U_media"]

Ua_media = df["Ua"].mean()
df
```

```python
df_colores = pd.read_csv("colores.csv")
df_colores["E_max_foton[eV]"] = df_colores["L_min[nm]"].apply(calcular_energia_foton).round(3)
df_colores["E_min_foton[eV]"] = df_colores["L_max[nm]"].apply(calcular_energia_foton).round(3)
df_colores["Uo"] = df_colores["E_min_foton[eV]"] - Ua_media
df_pwm = df_colores[["Color", "Uo"]].sort_values("Uo", ascending=False).copy()
df_colores
```

```python
df_pwm["Ganancia"] = df_pwm["Uo"] / 3.3
df_pwm["Division"] = (df_pwm["Ganancia"].max() / df_pwm["Ganancia"]).round(2)
df_pwm
```
